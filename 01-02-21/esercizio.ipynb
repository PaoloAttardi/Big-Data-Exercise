```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
print(df.shape[0])
print(df.isna().sum())
print(df['DEATH_EVENT'].value_counts())
sns.histplot(df, x='age')
```

```python
df_morti = df[df['DEATH_EVENT'] == 1]
((df_morti.groupby(['sex','DEATH_EVENT']).size())/(df_morti.shape[0])).unstack().plot.bar()
```

```python
df['age'] = pd.cut(df['age'], bins=5)
df.pivot_table(index=['age'], columns=['sex','smoking'])['DEATH_EVENT']
```

```python
df_anemia = df[df['anaemia'] == 1]
sns.histplot(df_anemia, x='diabetes', bins=2)
```
