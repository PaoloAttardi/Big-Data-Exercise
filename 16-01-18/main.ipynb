```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

df = pd.read_csv('bank.csv', sep=';')
print(df.shape[0])
print(df.isna().sum())

labels = df['y'].unique()
plt.pie(df['y'].value_counts(), labels=labels)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df['age'], color='orange')
plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()

temp = enc.fit_transform(df)
df_enc = pd.DataFrame(temp, columns=enc.get_feature_names_out())
df_enc = df_enc[['marital','y']]

corr = df_enc.corr()
sns.heatmap(corr, annot=True)
plt.show()

p1 = sns.barplot(df, x='marital', y=df_enc['y'], errwidth=0)
p1.bar_label(p1.containers[0])
plt.show()
```

```python
df_depo = df[df['y'] ==  'yes']
depo1 = df_depo.value_counts(normalize=True)

df_no_depo = df[df['y'] ==  'no']
depo0 = df_no_depo.value_counts(normalize=True)

sns.histplot(depo1, x='age', stat='percent')
plt.show()
sns.histplot(depo0, x='age', stat='percent')
plt.show()
```
